### 파일 압축풀기

```python
base_path = os.path.expanduser('~/aiffel/15_deep_learning/rock_scissors_paper/')
classes = ['rock', 'scissors', 'paper']

for i in classes:
    img_path = os.path.join(base_path, i)
    zip_file = os.path.join(img_path, f"{i}.zip")
    
    # Check if the zip file exists
    if os.path.exists(zip_file):
        os.chdir(img_path)
        
        # Unzip the file using ! command
        ! unzip -o {zip_file}
        
        print(f"Successfully unzipped {i}.zip to {img_path}")
    else:
        print(f"{zip_file} does not exist.")
```

### 이미지 사이즈 변환 

```python
target_size = (28, 28)

def resize_images(i):
    # 이미지 경로 설정
    img_path = base_path + f"/{i}/{i}"
    images=glob.glob(img_path + "/*.jpg")  
    print(len(images), "images to be resized.")

    # 이미지 사이즈 변환
    for img in images:
        old_image = Image.open(img)
        new_img = old_image.resize(target_size, Image.ANTIALIAS)
        new_img.save(img, "JPEG")
    print(len(images), "images resized")

for i in classes: 
    resize_images(i)
    
# Trial 1-4) 
# target_size=(28,28)
# 600 images to be resized.
# 600 images resized
# 600 images to be resized.
# 600 images resized
# 600 images to be resized.
# 600 images resized
```

### 데이터 불러오기

```python
# 바꿀 파라미터 설정
size_per_class=600
target_img_size=28
target_color=3  # 칼라 이미지

def load_data(base_path, number_of_data=size_per_class*len(classes)): 
    
    img_size=target_img_size
    color=target_color
    
    # 이미지 데이터와 라벨 데이터(가위: 0, 바위: 1, 보: 2)를 담을 행렬(matrix) 영역을 생성합니다
    imgs = np.zeros(number_of_data * img_size * img_size * color, dtype=np.int32).reshape(number_of_data, img_size, img_size, color)
    labels = np.zeros(number_of_data, dtype=np.int32)
    
    idx=0
    
    # 클래스별 레이블 붙이기
    for i in classes:  
        img_path = base_path + f"{i}/{i}"
        img_list = glob.glob(img_path + "/*.jpg")  
        for img in img_list:
            img = np.array(Image.open(img), dtype=np.int32)
            imgs[idx, :, :, :] = img  # 데이터 영역에 이미지 행렬을 복사

            if i == 'scissors':
                labels[idx] = 0
            elif i == 'rock':
                labels[idx] = 1
            elif i == 'paper':
                labels[idx] = 2

            idx = idx+1
        
    print("이미지 개수는", idx, '입니다.')
    return imgs, labels

# x, y에 데이터 로드하기
(x, y) = load_data(base_path)
print('x shape: {}'.format(x.shape))
print('y shape: {}'.format(y.shape))

# Trial 1-2) 
# img_size=28
# 학습데이터(x_train)의 이미지 개수는 300 입니다.
# x_train shape: (300, 28, 28, 3)
# y_train shape: (300,)

# Trial 2-4) 
# img_size=28
# 학습데이터(x_train)의 이미지 개수는 1800 입니다.
# x_train shape: (1800, 28, 28, 3)
# y_train shape: (1800,)
```

```python
plt.imshow(x[0])
```

### 입력값 정규화

```python
# 입력을 0-1 사이의 값으로 정규화
x_norm = x/255.0
```

### train-test split

```python
# statify를 사용해 train과 test에 각 클래스가 골고루 섞일 수 있도록 설정
x_train, x_test, y_train, y_test = train_test_split(
    x_norm, y, random_state=42, stratify=y)
```

```python
plt.imshow(x_train[0], cmap=plt.cm.binary)
print('라벨:', x_train[0])
```

## 딥러닝 네트워크 설계하기

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np

# 바꿔볼 수 있는 파라미터들
n_channel_1=32
n_channel_2=64
n_dense=64

# 딥러닝 네트워크 설계
model = keras.models.Sequential()
model.add(keras.layers.Conv2D(n_channel_1, (3,3), activation='relu', input_shape=(28, 28, 3)))
model.add(keras.layers.MaxPool2D(2,2))
model.add(keras.layers.Conv2D(n_channel_2, (3,3), activation='relu'))
model.add(keras.layers.MaxPool2D(2,2))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(n_dense, activation='relu'))
model.add(keras.layers.Dense(3, activation='softmax'))

# 서머리 출력
print('Model에 추가된 Layer 개수: ', len(model.layers))
model.summary()

# Trial 1-4) 
# n_channel_1=32
# n_channel_2=64
# n_dense=64
# Model에 추가된 Layer 개수:  7
# Model: "sequential_6"
# Model에 추가된 Layer 개수:  7
# Model: "sequential_6"
# _________________________________________________________________
# Layer (type)                 Output Shape              Param #   
# =================================================================
# conv2d_12 (Conv2D)           (None, 26, 26, 32)        896       
# _________________________________________________________________
# max_pooling2d_12 (MaxPooling (None, 13, 13, 32)        0         
# _________________________________________________________________
# conv2d_13 (Conv2D)           (None, 11, 11, 64)        18496     
# _________________________________________________________________
# max_pooling2d_13 (MaxPooling (None, 5, 5, 64)          0         
# _________________________________________________________________
# flatten_6 (Flatten)          (None, 1600)              0         
# _________________________________________________________________
# dense_12 (Dense)             (None, 64)                102464    
# _________________________________________________________________
# dense_13 (Dense)             (None, 3)                 195       
# =================================================================
# Total params: 122,051
# Trainable params: 122,051
# Non-trainable params: 0
```
